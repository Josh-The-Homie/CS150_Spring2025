```python
def triangle_type(a, b, c):
    if a == b == c:
        return "equilateral"
    elif a == b or b == c or a == c:
        return "isosceles"
    else:
        return "scalene"

# Get user input
a = float(input("Enter the length of side a: "))
b = float(input("Enter the length of side b: "))
c = float(input("Enter the length of side c: "))

# Check validity
if a <= 0 or b <= 0 or c <= 0:
    print("Invalid triangle: side lengths must be positive")
elif (a + b <= c) or (a + c <= b) or (b + c <= a):
    print("Invalid triangle: sum of two sides must exceed the third side")
else:
    triangle = triangle_type(a,b,c)
    print(f"The triangle is {triangle}")
```

```python
def analyze_grades():
    student_grades = []
    for i in range(5):
        while True:
            try:
                grade = float(input(f"Enter grade for student {i+1}: "))
                if 0 <= grade <= 100:
                    student_grades.append(grade)
                    break
                else:
                    print("Grade must be between 0 and 100. Please try again.")
            except ValueError:
                print("Invalid input. Please enter a numerical grade.")

    class_average = sum(student_grades) / len(student_grades)
    min_grade = min(student_grades)
    max_grade = max(student_grades)

    print(f"Class Average: {class_average:.2f}")
    print(f"Minimum Grade: {min_grade}")
    print(f"Maximum Grade: {max_grade}")

    at_risk_count = sum(1 for grade in student_grades if grade < 65)
    print(f"{at_risk_count} students are at risk in the course.")

analyze_grades()
```

```python
mult_of_3 = [i for i in range(3, 1001, 3)]
mult_of_4 = [i for i in range(4, 1001, 4)]

for num in mult_of_3:
  if num in mult_of_4:
    print(f"{num} is a multiple of 3 and 4")
  else:
    print(f"{num} is a multiple of 3 but not 4")
```

```python

def shopping_list_operations():
    shopping_list = []
    staple_foods = ["Rice", "Beans", "Corn", "Carrots", "Potatoes"]

    for i in range(7):
        item = input(f"Enter item {i+1} (or 'done' to finish): ")
        if item.lower() == "done":
            break
        shopping_list.append(item)

    print(f"\nTotal items: {len(shopping_list)}")

    print("\nReversed list:", shopping_list[::-1])

    shopping_list.sort()
    print("\nAlphabetical list:", shopping_list)

    staple_count = 0
    for food in staple_foods:
        if food in shopping_list:
            staple_count += 1

    print(f"\n{staple_count} staple foods are present in your list.")

shopping_list_operations()
```
